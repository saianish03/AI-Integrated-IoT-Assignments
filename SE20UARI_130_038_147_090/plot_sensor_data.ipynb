```python
import json
import matplotlib.pyplot as plt
```

```python
file_path = "DHT_sensor_data.json"
with open(file_path, 'r') as file:
    data = json.load(file)
```

```python
humidity_values = []
temperature_values = []

for key, value in data["Status"].items():
    humidity_values.append(value["Humidity"])
    temperature_values.append(value["Temperature"])
```

```python
plt.figure(figsize=(12, 6))

# Humidity Plot
plt.subplot(2, 1, 1)
plt.plot(humidity_values, label='Humidity', color='blue')
plt.title('Humidity Over Time')
plt.xlabel('Time')
plt.ylabel('Humidity')
plt.legend()

# Temperature Plot
plt.subplot(2, 1, 2)
plt.plot(temperature_values, label='Temperature', color='red')
plt.title('Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()

plt.tight_layout()

plt.show()
```
